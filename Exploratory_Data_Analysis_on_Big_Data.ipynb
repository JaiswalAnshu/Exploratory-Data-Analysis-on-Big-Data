# DATA CLEANING

```python
# importing the required libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt #visualizing data
%matplotlib inline
import seaborn as sns
```

```python
# importing the csv  file 
# used 'unicode_escape' to avoid encoding error
df=pd.read_csv("Diwali Sales Data.csv" , encoding='unicode_escape')
```

```python
# Finding total number of Rows and columns in csv file
df.shape
```

```python
# Getting the details of top Five Rows
df.head(8)
```

```python
# Getting the details of bottom Ten Rows
df.tail(10)
```

```python
# Getting details of all the columns in csv file 
df.info()
```

```python
# Droping the null columns 
df.drop(['Status'],['unnamed1'] , axis =1 , inplace= True)
```

```python
# Finding the total number of columns after droping the null columns 
df.shape
```

```python
# Finding the column that contains null values
df.info()
```

```python
# finding the sum of all the null values in csv file in each column 
pd.isnull(df).sum()
```

```python
# Droping the null values in Table
df.dropna(inplace=True)
```

```python
# Finding if the null values  deleted or not 
df.shape
```

```python
# Finding if any null values exists or not 
pd.isnull(df).sum()
```

```python
# change data type of a column 
df['Amount'] = df['Amount'].astype('int')
```

```python
# Finding the data type of renewed Amount column 
df['Amount'].dtype
```

```python
df.columns
```

```python
df.rename(columns = {'Marital_Status' : 'Shaddi'})
```

# EXPLORATORY DATA ANALYSIS

GENDER

```python
ax= sns.countplot(x= 'Gender' , data = df )

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
 Sales_gen= df.groupby(['Gender'] , as_index = False) ['Amount'].sum().sort_values(by='Amount' , ascending = False)

sns.barplot(x= 'Gender' , y = 'Amount' , data = Sales_gen )
```

From the above graphs we see that the purchase done by females are much higher than the purchase done by male and also the amount spent by the females are higher . 

# Age

```python
df.columns
```

```python
bx=sns.countplot(x='Age Group' , data = df , hue = 'Gender')

for bars in bx.containers:
    bx.bar_label(bars)
```

```python
Sales_age= df.groupby(['Age Group'] , as_index = False) ['Amount'].sum().sort_values(by='Amount' , ascending = False)

sns.barplot(x= 'Age Group' , y = 'Amount' , data = Sales_age )
```

From the above graphs we see that most of the buyers are of age group between 26-35 years feamle .  

# State

```python
df.columns
```

```python
Order_state = df.groupby(['State'] , as_index = False) ['Orders'].sum().sort_values(by='Orders' , ascending = False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x= 'State' , y = 'Orders' , data = Order_state )
```

```python
Order_state = df.groupby(['State'] , as_index = False) ['Amount'].sum().sort_values(by='Amount' , ascending = False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x= 'State' , y = 'Amount' , data = Order_state )
```

From the above graphs we can see that most of the orders are from uttar pradesh , maharastra and karnataka and also the purchasing power of these three states are higher .
